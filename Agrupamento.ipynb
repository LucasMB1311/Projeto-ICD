```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans

diet_df = pd.read_csv("diet.csv", encoding = "UTF-8")
```

```python
diet_df
```

```python
diet_group = diet_df[["Carbohydrates (g)", "Total Sugar (g)"]]
```

```python
diet_group.head(5)
```

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans.fit(diet_group)
```

```python
kmeans.labels_
```

```python
sns.scatterplot(data=diet_group, x="Carbohydrates (g)", y="Total Sugar (g)", hue=kmeans.labels_)
```

```python
diet_df[kmeans.labels_ == 0]
```

```python
diet_df[kmeans.labels_ == 1]
```
